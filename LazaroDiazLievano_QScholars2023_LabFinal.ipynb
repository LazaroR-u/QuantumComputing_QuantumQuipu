$$CU3(\theta,\phi, \lambda)= I \otimes \ket{0} \bra{0} + U3(\theta, \phi,\lambda) \otimes \ket{1} \bra{1} $$

donde $I$ es la matriz identidad y $U3$ la compuerta U3,

$$ U3(\theta, \phi, \lambda ) =
\begin{pmatrix}
\cos{\frac{\theta}{2}} & -e^{i\lambda}\sin{\frac{\theta}{2}} \\
e^{i\phi}\sin{\frac{\theta}{2}} & e^{i(\phi +\lambda)}\cos{\frac{\theta}{2}}
\end{pmatrix}, \quad \quad I = \begin{pmatrix} 1 & 1 \\ 1 & 1 \end{pmatrix} $$

Expresando los kets y bras como vectores, podemos realizar los productos tensoriales correspondientes y escribir la compuerta CU3 en su forma matricial.

$$ U3(\theta, \phi, \lambda ) =
\begin{pmatrix}
1 & 0 & 0 & 0 \\
0 & \cos{\frac{\theta}{2}} & 0 & -e^{i\lambda}\sin{\frac{\theta}{2}} \\
0 & 0 & 1 & 0 \\
0 & e^{i\phi}\sin{\frac{\theta}{2}} & 0 &  e^{i(\phi +\lambda)}\cos{\frac{\theta}{2}}

## 4.- A partir de la relación existente entre la operación (módulo 2) y el operador XOR ($\oplus$). Mostrar:

\begin{align}
\sum_{i=1}^n x_iz_i \pmod 2=x_1z_1\oplus x_2z_2\oplus\cdots\oplus x_nz_n
\end{align}


Para demostrar esta relación, podemos observar que tanto la suma módulo 2 como la operación XOR siguen las mismas reglas de adición binaria. En ambos casos, la suma de dos valores binarios resulta en 0 si ambos valores son iguales y en 1 si son diferentes.


La relacion existente entre la funcion XOR y el modulo 2 es la siguiente

$$ a \oplus b = (a+b) \pmod 2 $$

Ahora, consideremos la expansión de la suma en la parte izquierda de la ecuación:

\begin{align}
\sum_{i=1}^n x_iz_i \pmod 2 &= (x_1z_1 + x_2z_2 + \cdots + x_nz_n) \pmod 2 \\
&= (x_1z_1 \pmod 2 + x_2z_2 \pmod 2 + \cdots + x_nz_n \pmod 2) \pmod 2
\end{align}

Debido a que estamos tomando el módulo 2 en cada término individual $x_iz_i$, los valores pares se convierten en 0 y los valores impares se convierten en 1. Por lo tanto, la expansión se convierte en:

\begin{align}
&= (x_1z_1 + x_2z_2 + \cdots + x_nz_n) \pmod 2 \\
&= (x_1z_1\oplus x_2z_2\oplus\cdots\oplus x_nz_n)
\end{align}

Esto demuestra que la expresión en la parte izquierda de la ecuación es equivalente a la expresión en la parte derecha de la ecuación, como queríamos demostrar:

\begin{align}
\sum_{i=1}^n x_iz_i \pmod 2 = x_1z_1\oplus x_2z_2\oplus\cdots\oplus x_nz_n
\end{align}

En resumen, esta relación muestra cómo la operación de suma de productos módulo 2 es equivalente a la operación XOR de los productos individuales $x_i z_i$. Esto es útil en el contexto de la computación cuántica y los cálculos con bits cuánticos (qubits)


## 6.- En la construcción más simple posible de un oráculo balanceado, se utiliza la compuerta $CX$, la cual toma como objetivo el $(n+1)$-ésimo qubit del circuito y como control cada uno de los $n$ primeros qubits. Sin embargo, ¿es realmente necesario aplicar $CX$ a todos los $n$ primeros qubits, o podríamos prescindir de algunos y obtener igualmente un oráculo balanceado?


La compuerta CX (CNOT) en el oráculo balanceado se aplica como una puerta controlada, donde los primeros n qubits son los qubits de control y el (n+1)-ésimo qubit es el qubit objetivo. La compuerta CX se activará (invertirá el valor del qubit objetivo) solo si todos los qubits de control están en el estado ∣1⟩.
En la f(x), solo hay cuatro combinaciones en las que el resultado es 1:
\begin{align}
f(011)&=1\\
f(010)&=1\\
f(100)&=1\\
f(101)&=1\\
\end{align}


Esto significa que si aplicamos una compuerta CX a estos cuatro estados en particular, el oráculo será balanceado. No es necesario aplicar la compuerta CX a todos los n primeros qubits; solo necesitas aplicarla a aquellos que corresponden a las combinaciones que tienen resultado 1.

Por lo tanto, en lugar de aplicar la compuerta CX a todos los n primeros qubits, se pueden aplicar solo a los tres primeros qubits correspondientes a las combinaciones 011, 010, 100 y 101.


Esto aún lograría la propiedad deseada de tener un oráculo balanceado.


```python
oraculo_c = QuantumCircuit(4)

oraculo_c.x(1)
oraculo_c.barrier()

#aplicamos la compuerta cx solamente al primer y tercer qubit, es decir,
#el segundo qubit no cuenta con la compuerta cx
oraculo_c.cx(1,3)
oraculo_c.cx(2,3)

oraculo_c.barrier()

oraculo_c.x(1)
oraculo_c.barrier()

qcc2=QuantumCircuit(4,3)
qcc2.x(3)
qcc2.h(3)
qcc2.barrier()

for x in range(3):
    qcc.h(x)
qcc2.barrier()

qcc2=qcc2.compose(oraculo_c)
qcc2.barrier

for x in range(3):
    qcc2.h(x)
qcc2.barrier()

#qc.measure(range(4),range(4))
qcc2.measure([0,1,2],[0,1,2])
qcc2.draw()
```

```python
aer_c = Aer.get_backend('aer_simulator')
result = aer_c.run(qcc2).result()
answe = result.get_counts()

plot_histogram(answe)
```

```python
def oracle_part(qc,qr):
    mcx = XGate().control(3)
    mccx = XGate().control(4)
    #Los qubits del 0 al 4 son los nodos.
    #Los qubits del 5 al 10 nos permiten reconocer si la arista es correcta
    #(i.e. si se da entre nodos de distinto color).

    #estas combinaciones no son posibles dado que no existen aristas entre estos nodos
    #estos solo son dos ejemplos de aristas no posibles, existen otros que no se escribieron.
    #qc.cx(qr[0], qr[5])
    #qc.cx(qr[1], qr[5])
    #qc.barrier()

    #qc.cx(qr[0], qr[6])
    #qc.cx(qr[2], qr[6])
    #qc.barrier()

    #arista 0-3
    qc.cx(qr[0], qr[5])
    qc.cx(qr[3], qr[5])
    qc.barrier()
    #arista 0-4
    qc.cx(qr[0], qr[6])
    qc.cx(qr[4], qr[6])
    qc.barrier()

    #arista 1-3
    qc.cx(qr[1], qr[7])
    qc.cx(qr[3], qr[7])
    qc.barrier()
    #arista 1-4
    qc.cx(qr[1], qr[8])
    qc.cx(qr[4], qr[8])
    qc.barrier()

    #arista 2-3
    qc.cx(qr[2], qr[9])
    qc.cx(qr[3], qr[9])
    qc.barrier()
    #arista 2-4
    qc.cx(qr[2], qr[10])
    qc.cx(qr[4], qr[10])
    qc.barrier()

    #con esto ya escribimos las 5 aristas posibles de nuestro sistema.

    #Los qubits del 11 al 14 nos van a permitir contar la cantidad de aristas correctas
    #de tal forma que la suma será expresada en binario con cifras dadas por q_11, q_12, q_13 y q_14.
    qc.barrier()
    qc.cx(qr[5],qr[11])
    qc.barrier()

    qc.ccx(qr[6], qr[11], qr[12])
    qc.cx(qr[6],qr[11])
    qc.barrier()

    qc.ccx(qr[7], qr[11], qr[12])
    qc.cx(qr[7],qr[11])
    qc.barrier()

    qc.append(mcx,[8,11,12,13])
    qc.ccx(qr[8], qr[11], qr[12])
    qc.cx(qr[8],qr[11])
    qc.barrier()

    qc.append(mcx,[9,11,12,13])
    qc.ccx(qr[9], qr[11], qr[12])
    qc.cx(qr[9],qr[11])
    qc.barrier()

    qc.append(mccx,[10,11,12,13,14])
    qc.append(mcx, [10,11,12,13])
    qc.ccx(qr[10], qr[11], qr[12])
    qc.cx(qr[10],qr[11])
    qc.barrier()

    qc.x(qr[11])
    qc.x(qr[12])
    qc.x(qr[13])
    qc.barrier()

    #El qubit 15 indicará cuando se haya encontrado un estado deseado
    qc.append(mccx,[11,12,13,14,15])

    return qc
```

```python
qr=QuantumRegister(16)
qc=QuantumCircuit(qr)
oracle_part(qc,qr).draw()
```

```python
def oracle(qc,qr):

    qc=oracle_part(qc,qr)
    qc2=qc.inverse()
    qc.z(15)
    qc.compose(qc2,inplace=True)

    return qc
```

```python
qr=QuantumRegister(16)
qc=QuantumCircuit(qr)
oracle(qc,qr).draw()
```

```python
def grover_diffusion(qc,qr):
    mcz = ZGate().control(4)

    qc.barrier()
    qc.h(range(5))
    qc.x(range(5))
    qc.append(mcz,[0,1,2,3,4])
    qc.x(range(5))
    qc.h(range(5))
    qc.barrier()
    return qc
```

```python
def grover(qc,qr):
    return grover_diffusion(oracle(qc,qr),qr)
```

\ket{s}\bra{s}=\sin^2{\theta}\ket{A}\bra{A}+\cos^2{\theta}\ket{B}\bra{B}+\sin{\theta}\cos{\theta}\left[\ket{A}\bra{B}+\ket{B}\bra{A}\right]
\end{align}

\\

con esta expresion, ahora lo sustituimos en la ecuacion de $\ket{\psi_3}$


\begin{align}
\ket{\psi_3}&=\left(2\ket{s}\bra{s}-\mathbb{I}\right)\left(-\sqrt{\frac{a}{2^n}}\ket{A}+\sqrt{\frac{b}{2^n}}\ket{B}\right) \\
&= \left(2\left[\sin^2{\theta}\ket{A}\bra{A}+\cos^2{\theta}\ket{B}\bra{B}+\sin{\theta}\cos{\theta}\left[\ket{A}\bra{B}+\ket{B}\bra{A}\right] \right]-\mathbb{I}\right)\left(-\sin{\theta}\ket{A}+\cos{\theta}\ket{B}\right)\\
&= 2\sin^2{\theta}\ket{A}\bra{A}+\cos^2{\theta}\ket{B}\bra{B}+\sin{\theta}\cos{\theta}\left[\ket{A}\bra{B}+\ket{B}\bra{A}\right] \\
&= -2\sin^3{\theta}\ket{A}\braket{A}{A} + 2\sin^2{\theta}\cos{\theta}\ket{A}\underbrace{\braket{A}{B}} - 2\cos^2{\theta}\sin{\theta}\ket{B}\underbrace{\braket{B}{A}}+2\cos^3{\theta}\ket{B}\braket{B}{B}\\
& + 2\sin{\theta}\cos{\theta}\left[-\sin{\theta}\ket{A}\underbrace{\braket{B}{A}} +\cos{\theta}\ket{A}\braket{B}{B} -\sin{\theta}\ket{B}\braket{A}{A}+\cos{\theta}\ket{B}\underbrace{\braket{A}{B}}\right]
\end{align}

\\
usaremos la condicion de ortonormalidad de las bases, esto es $\braket{A}{A}=1$ y $\braket{A}{B}=0$, entonces nos queda lo siguiente

\
\begin{align}
\ket{\psi_3} &= \left[-2\sin^3{\theta} + 2\cos^2{\theta}\sin{\theta}+\sin{\theta}\right]\ket{A} + \left[ 2\cos^3{\theta}-2\cos{\theta}\sin^2{\theta}-\cos{\theta}
  \right] \ket{B} \\
&=  \left[-2\sin^3{\theta} + 2(1-\sin^2{\theta})\sin{\theta}+\sin{\theta}\right]\ket{A} + \left[ 2\cos^3{\theta}-2\cos{\theta}(1-\cos^2{\theta})-\cos{\theta}
  \right] \ket{B}  \\
  &= \left[  3\sin{\theta}-4\sin^3{\theta}  \right]\ket{A} +
  \left[ 4\cos^3{\theta}-3\cos{\theta} \right] \ket{B} \\
  &= \sin3\theta \ket{A} + \cos{3\theta} \ket{B}
\end{align}

```python
import numpy as np
from numpy import pi
# importing Qiskit
from qiskit import QuantumCircuit, transpile, Aer, IBMQ
from qiskit.providers.ibmq import least_busy
from qiskit.tools.monitor import job_monitor
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

```python
def qft_rotations(circuit, n):
    if n == 0: # Exit function if circuit is empty
        return circuit
    n -= 1 # Indexes start from 0
    circuit.h(n) # Apply the H-gate to the most significant qubit
    for qubit in range(n):
        # For each less significant qubit, we need to do a
        # smaller-angled controlled rotation:
        circuit.cp(pi/2**(n-qubit), qubit, n)

# Let's see how it looks:
qc = QuantumCircuit(4)
qft_rotations(qc,4)
qc.draw(output='mpl')
```

```python
def qft_rotations(circuit, n):
    """Performs qft on the first n qubits in circuit (without swaps)"""
    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cp(pi/2**(n-qubit), qubit, n)
    # At the end of our function, we call the same function again on
    # the next qubits (we reduced n by one earlier in the function)
    qft_rotations(circuit, n)

# Let's see how it looks:
qc = QuantumCircuit(4)
qft_rotations(qc,4)
qc.draw(output='mpl')
```

```python
def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    """QFT on the first n qubits in circuit"""
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit

# Let's see how it looks:
qc = QuantumCircuit(4)
qft(qc,4)
qc.draw(output='mpl')
```

```python
qft(qc,3)
qc.draw(output='mpl')
```

```python
def inverse_qft(circuit, n):
    """Does the inverse QFT on the first n qubits in circuit"""
    # First we create a QFT circuit of the correct size:
    qft_circ = qft(QuantumCircuit(n), n)
    # Then we take the inverse of this circuit
    invqft_circ = qft_circ.inverse()
    # And add it to the first n qubits in our existing circuit
    circuit.append(invqft_circ, circuit.qubits[:n])
    return circuit.decompose() # .decompose() allows us to see the individual gates
```

```python
nqubits= 3

number = 4
qc = QuantumCircuit(nqubits)
for qubit in range(nqubits):
    qc.h(qubit)
qc.p(number*pi/4,0)
qc.p(number*pi/2,1)
qc.p(number*pi,2)

qc.draw()
```

```python
qc_init = qc.copy()
qc_init.save_statevector()
sim = Aer.get_backend("aer_simulator")
statevector = sim.run(qc_init).result().get_statevector()
plot_bloch_multivector(statevector)
```

```python
qc = inverse_qft(qc, nqubits)
qc.measure_all()
qc.draw()
```
